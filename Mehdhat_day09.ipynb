```python
import seaborn as sns
import matplotlib.pyplot as plt

my_data = sns.load_dataset("tips")

my_analysis = sns.barplot(
    x='time',
    y='day',
    data=my_data,
    hue='sex',
    hue_order=['Male', 'Female'],
    orient='h',
    capsize=0.2,
    palette='pastel'
)

my_analysis.set_title("Tips by Day and Gender", fontsize=16)
my_analysis.set_xlabel("Time", fontsize=12)
my_analysis.set_ylabel("Days", fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = sns.load_dataset('titanic')

data_cleaned = data.dropna(subset=['age', 'fare'])

plt.figure(figsize=(10, 8))
sns.pairplot(data_cleaned, hue="sex", hue_order=['male', 'female'], markers=['o', '*'], kind='kde', palette='pastel')

plt.suptitle('Pair Plot of Titanic Dataset (Age and Fare)', fontsize=16)

plt.show()
```
