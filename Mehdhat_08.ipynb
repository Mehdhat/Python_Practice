```python
import matplotlib.pyplot as plt
# **Area plot/ Stack plot**
days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]
plt.plot([],[],color='g', label='Sleeping', linewidth=7)
plt.plot([],[],color='c', label='Eating', linewidth=7)
plt.plot([],[],color='y', label='Working', linewidth=7)
plt.plot([],[],color='w', label='Playing', linewidth=7)
plt.stackplot(days, sleeping,eating,working,playing, colors=['g','c','y','w'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area Plot')
plt.legend()
plt.show()
```
