```python
#TASK 01

#Create a Pandas DataFrame with the following columns: 'Product', 'Category', 'Sales_Q1', 'Sales_Q2', 'Sales_Q3', 'Sales_Q4'.
#Populate the DataFrame with data for 8 different products.

import pandas as pd

data = {'Product': ['1', '2', '3', '4', '5', '6', '7', '8'],
        'Category': ['UTILITIES', 'CLOTHES', 'GROCERIES', 'RENT', 'TRAVEL', 'ENTERTAINMENT', 'MEDICAL', 'EDUCATION'],
        'Sales_Q1': [16000, 2500, 20000, 12000, 5000, 3000, 4000, 10000],
        'Sales_Q2': [1020, 1700, 2200, 1040, 2500, 1500, 2000, 5000],
        'Sales_Q3': [2400, 1090, 2400, 1660, 3000, 1000, 1500, 6000],
        'Sales_Q4': [1760, 7210, 2560, 1800, 3500, 2000, 2500, 7000]}

df = pd.DataFrame(data)
print(df)
```

```python
# Add a new column 'Total_Sales'
print("Total Sales")
df['Total_Sales'] = df['Sales_Q1'] + df['Sales_Q2'] + df['Sales_Q3'] + df['Sales_Q4']
print(df['Total_Sales'])
```

```python
# Add a new column 'Average_Sales'
print("Average Sales")
df['Average_Sales'] = df['Total_Sales'] / 4
print(df['Average_Sales'])
```

```python
# Group the data by 'Category' and calculate the total sales for each category
category_sales = df.groupby('Category')['Total_Sales'].sum()
print(category_sales)
```

```python
# Find the category with the highest total sales
highest_sales_category = category_sales.idxmax()
print("Category with highest total sales:", highest_sales_category)
```

```python
# Create a pie chart showing the proportion of total sales for each category.

import matplotlib.pyplot as plt

# Pie chart
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%')
plt.title('Proportion of Total Sales by Category')
plt.show()
```

```python
# Create a stacked bar plot showing sales for each quarter for each product.

df[['Sales_Q1', 'Sales_Q2', 'Sales_Q3', 'Sales_Q4']].plot(kind='bar', stacked=True)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Quarterly Sales by Product')
plt.legend(title='Quarter')
plt.show()
```

```python
# Scatter plot
plt.scatter(df['Total_Sales'], df['Average_Sales'])
plt.xlabel('Total Sales')
plt.ylabel('Average Sales')
plt.title('Total Sales vs Average Sales')
plt.show()
```
