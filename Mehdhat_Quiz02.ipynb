```python
#import dataset from github using seaborn
import seaborn as sns
my_data=sns.load_dataset("iris")
my_data.head(10)
```

```python
#Load the Iris dataset into a pandas DataFrame
import pandas as pd
df=pd.DataFrame(my_data)
df.head(10)
df.to_excel("/content/drive/MyDrive/Files/iris.xlsx", index=False)
df.to_csv("/content/drive/MyDrive/Files/iris.csv", index=False)
```

```python
# Check for missing values
missing_values = my_data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
#Create a histogram for petal lengths
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.hist(my_data['petal_length'], bins=10, color='#FF69B4', edgecolor='black', alpha=0.7)

plt.xlabel('Petal Length (cm)', fontsize=14)
plt.ylabel('Number of Occurrences', fontsize=14)
plt.title('Petal Length Distribution in the Iris Dataset', fontsize=16)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
#Plot a scatter plot of sepal length vs. sepal width
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))

plt.scatter(my_data['sepal_length'], my_data['sepal_width'],
            c='#AB1546', edgecolor='black', alpha=0.7, s=100)  # s controls the size of points

plt.xlabel('Sepal Length (cm)', fontsize=14)
plt.ylabel('Sepal Width (cm)', fontsize=14)
plt.title('Sepal Length vs. Sepal Width in the Iris Dataset', fontsize=16)

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
#Generate a line plot for petal length trends (if grouping by species)
import matplotlib.pyplot as plt

grouped_data = my_data.groupby('species')['petal_length'].mean()

plt.figure(figsize=(10, 6))

grouped_data.plot(kind='line', marker='o', color='purple', linewidth=2, markersize=8)

plt.xlabel('Species', fontsize=14)
plt.ylabel('Mean Petal Length (cm)', fontsize=14)
plt.title('Mean Petal Length by Species', fontsize=16)

plt.xticks(rotation=45)

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
#Calculate the correlation coefficient between sepal length and sepal width

correlation_coefficient = my_data['sepal_length'].corr(my_data['sepal_width'])
print("Correlation coefficient between sepal length and sepal width:", correlation_coefficient)
```

```python
#Create a bar chart showing the average petal length for each species.
import matplotlib.pyplot as plt

grouped_data = my_data.groupby('species')['petal_length'].mean()

plt.figure(figsize=(10, 6))

colors = ['#FF69B4', '#32CD32', '#00CED1']
grouped_data.plot(kind='bar', color=colors, edgecolor='black')

plt.xlabel('Species', fontsize=14)
plt.ylabel('Average Petal Length (cm)', fontsize=14)
plt.title('Average Petal Length by Species', fontsize=16)

plt.xticks(rotation=0)

plt.grid(axis='y', linestyle='--', alpha=0.6)

for index, value in enumerate(grouped_data):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```
